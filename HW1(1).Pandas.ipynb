## 9. Определить статистику заказов стейков, а также статистику заказов стейков с соусами той или иной остроты.

### Статистика по стейкам

### Статистика по стейкам с соусами той или иной степени остроты в том или ином количестве

**Общая статистика по всем заказам со стейками с соусами той или иной степени остроты в том или ином количестве**

```python
stakesMildMediumHot = df[df.item_name.str.contains("Steak") 
                         & df.choice_description.str.contains("Mild|Medium|Hot", na=False)]
stakesMildMediumHot.loc[:,'sause_sharpness'] = ""
stakesMildMediumHot.loc[:,'sause_sharpness'] = stakesMildMediumHot.loc[:,'sause_sharpness'].astype('object')
i = 0
for index,row in stakesMildMediumHot.iterrows():
    stakesMildMediumHot.at[index,'sause_sharpness'] = re.findall('\(([^)]+)', stakesMildMediumHot.loc[:,'choice_description'].values[i])
    stakesMildMediumHot.loc[index,'sause_sharpness'].sort()
    i += 1
stakesMildMediumHot.loc[:,'sause_sharpness'] = stakesMildMediumHot.loc[:,'sause_sharpness'].astype('str')
stakesMildMediumHot.loc[:,'sause_sharpness'] = stakesMildMediumHot.sause_sharpness.str.strip("[]")
stakesMildMediumHot.loc[:,'sause_sharpness'] = stakesMildMediumHot.sause_sharpness.str.replace("'", "")
stakesMildMediumHot.groupby('sause_sharpness')[["quantity", "item_price"]].describe()
```

**Статистика заказов различных стейков с соусами той или иной степени остроты в том или ином количестве**

```python
stakesMildMediumHot = df[df.item_name.str.contains("Steak") 
                         & df.choice_description.str.contains("Mild|Medium|Hot", na=False)]
stakesMildMediumHot.loc[:,'sause_sharpness'] = ""
stakesMildMediumHot.loc[:,'sause_sharpness'] = stakesMildMediumHot.loc[:,'sause_sharpness'].astype('object')
i = 0
for index,row in stakesMildMediumHot.iterrows():
    stakesMildMediumHot.at[index,'sause_sharpness'] = re.findall('\(([^)]+)', stakesMildMediumHot.loc[:,'choice_description'].values[i])
    stakesMildMediumHot.loc[index,'sause_sharpness'].sort()
    i += 1
stakesMildMediumHot.loc[:,'sause_sharpness'] = stakesMildMediumHot.loc[:,'sause_sharpness'].astype('str')
stakesMildMediumHot.loc[:,'sause_sharpness'] = stakesMildMediumHot.sause_sharpness.str.strip("[]")
stakesMildMediumHot.loc[:,'sause_sharpness'] = stakesMildMediumHot.sause_sharpness.str.replace("'", "")
stakesMildMediumHot.groupby(['item_name','sause_sharpness'])[["quantity", "item_price"]].describe()
```
